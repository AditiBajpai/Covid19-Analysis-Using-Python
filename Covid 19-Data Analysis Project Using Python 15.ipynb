Filter the dataframe with only this columns
   ['continent','location','date','total_cases','total_deaths','gdp_per_capita','
   human_development_index'] and update the data frame.

```python
data = data.loc[:,['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']]
```

```python
data.head()
```

#  Data Cleaning

Remove all duplicates observations

```python
data.drop_duplicates(inplace=True)
```

Find missing values in all columns

```python
data.isnull().sum()
```

Remove all observations where continent column value is missing

```python
data.dropna(subset=['continent'],inplace=True)
```

Fill all missing values with 0

```python
data.fillna(value=0,inplace=True)
```

# Date time format 

Convert date column in datetime format using pandas.to_datetime

```python
data['date'] = pd.to_datetime(data['date'])
```

Create new column month after extracting month data from date column.

```python

data['month'] = pd.to_datetime(data['date']).dt.month
```

```python
data
```

# Data Aggregation

Find max value in all columns using groupby function on 'continent'
        column

```python
data.groupby('continent').max()
```

Store the result in a new dataframe named 'df_groupby'.

```python
df_groupby = data.reset_index()
```

```python
df_groupby.head(10)
```

# Feature Engineering

Create a new feature 'total_deaths_to_total_cases' by ratio of
        'total_deaths' column to 'total_cases'

```python
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths'] / df_groupby['total_cases']
```

```python
df_groupby.head()
```

# Data Visualization

Perform Univariate analysis on 'gdp_per_capita' column by plotting
        histogram using seaborn dist plot.

```python
sns.distplot(df_groupby['gdp_per_capita'],kde=False,bins=50)
```

Plot a scatter plot of 'total_cases' & 'gdp_per_capita'

```python
sns.scatterplot(x = 'total_cases',y = 'gdp_per_capita',data=df_groupby)
```

 Plot Pairplot on df_groupby dataset.

```python
sns.pairplot(df_groupby)
```

Plot a bar plot of 'continent' column with 'total_cases' 

```python
sns.catplot(x = 'continent',y = 'total_cases',kind = 'bar',data = df_groupby)
```

Save the df_groupby dataframe in your local drive using pandas.to_csv
